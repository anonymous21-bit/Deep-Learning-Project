```python
import os
import warnings
import math
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, regularizers, optimizers
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
import sys

warnings.filterwarnings("ignore", message="resource_tracker:")

# TPU setup (commented out for CPU debugging)
# try:
#     tpu = tf.distribute.cluster_resolver.TPUClusterResolver()
#     tf.config.experimental_connect_to_cluster(tpu)
#     tf.tpu.experimental.initialize_tpu_system(tpu)
#     strategy = tf.distribute.TPUStrategy(tpu)
#     print("Running on TPU:", tpu.master())
# except ValueError:
strategy = tf.distribute.get_strategy()
print("Running on CPU/GPU")

# Kaggle dataset download
!pip install -q kaggle
from google.colab import files
files.upload()  # Upload kaggle.json
!mkdir -p ~/.kaggle
!mv kaggle.json ~/.kaggle/
!chmod 600 ~/.kaggle/kaggle.json
!kaggle datasets download -d zongaobian/h1b-lca-disclosure-data-2020-2024 -f Combined_LCA_Disclosure_Data_FY2020_to_FY2024.csv
DATA_PATH = "Combined_LCA_Disclosure_Data_FY2020_to_FY2024.csv"

# Global settings
SAMPLE_SIZE = 500000
VAL_SPLIT = 0.3
TOTAL_ROWS = 3564698
VAL_ROWS = int(TOTAL_ROWS * VAL_SPLIT)
TRAIN_ROWS = TOTAL_ROWS - VAL_ROWS
BATCH_SIZE = 8192
TRAIN_STEPS = math.ceil(TRAIN_ROWS / BATCH_SIZE)
VAL_STEPS = math.ceil(VAL_ROWS / BATCH_SIZE)
DEBUG_STEPS = 10  # Temporary for debugging

print(f"Training steps per epoch: {TRAIN_STEPS}")
print(f"Validation steps per epoch: {VAL_STEPS}")
print(f"Debug steps per epoch: {DEBUG_STEPS}")

# Helper Functions
def drop_high_missing(df, threshold=0.5):
    missing_ratio = df.isnull().mean()
    cols_to_drop = missing_ratio[missing_ratio > threshold].index.tolist()
    if cols_to_drop:
        print("Dropping columns due to high missing values:", cols_to_drop)
        df = df.drop(columns=cols_to_drop)
    return df

def impute_missing(df):
    num_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
    cat_cols = df.select_dtypes(include=['object']).columns.tolist()
    df[num_cols] = df[num_cols].fillna(df[num_cols].median())
    df[cat_cols] = df[cat_cols].fillna("Missing")
    return df

def remove_high_corr(df, corr_threshold=0.95):
    num_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
    corr_matrix = df[num_cols].corr().abs()
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    to_drop = [column for column in upper.columns if any(upper[column] > corr_threshold)]
    if to_drop:
        print("Dropping highly correlated numeric columns:", to_drop)
        df = df.drop(columns=to_drop)
    return df

def filter_categorical_features(df, max_unique=50):
    cat_cols = df.select_dtypes(include=['object']).columns.tolist()
    low_card = [col for col in cat_cols if df[col].nunique() <= max_unique]
    high_card = [col for col in cat_cols if df[col].nunique() > max_unique]
    print("Low-cardinality features:", low_card)
    print("High-cardinality features:", high_card)
    return low_card, high_card

# Custom Frequency Encoder
class FrequencyEncoder:
    def __init__(self):
        self.freq_maps = {}
        self.feature_names = None

    def fit(self, X, y=None):
        if isinstance(X, np.ndarray):
            X = pd.DataFrame(X, columns=[f"col{i}" for i in range(X.shape[1])])
        self.feature_names = X.columns.tolist()
        for col in X.columns:
            self.freq_maps[col] = X[col].value_counts(normalize=True).to_dict()
        return self

    def transform(self, X):
        if isinstance(X, np.ndarray):
            X = pd.DataFrame(X, columns=self.feature_names)
        X_enc = X.copy()
        for col in X_enc.columns:
            X_enc[col] = X_enc[col].map(self.freq_maps.get(col, {})).fillna(0)
        return X_enc.values

    def fit_transform(self, X, y=None):
        return self.fit(X, y).transform(X)

    def get_feature_names_out(self, input_features=None):
        if self.feature_names is None:
            raise ValueError("FrequencyEncoder must be fitted before calling get_feature_names_out")
        return np.array(self.feature_names if input_features is None else input_features)

# Load and preprocess sample data
print(f"Loading {SAMPLE_SIZE}-row sample for preprocessing...")
df_sample = pd.read_csv(DATA_PATH, nrows=SAMPLE_SIZE, low_memory=False)

categories = ['Certified', 'Certified - Withdrawn', 'Denied', 'Withdrawn']
df_sample['target'] = pd.Categorical(df_sample['CASE_STATUS'], categories=categories).codes
df_sample = df_sample.drop(columns=['CASE_STATUS', 'CASE_NUMBER'])

date_columns = ['RECEIVED_DATE', 'DECISION_DATE', 'ORIGINAL_CERT_DATE', 'BEGIN_DATE', 'END_DATE']
for col in date_columns:
    df_sample[col] = pd.to_datetime(df_sample[col], errors='coerce')
    df_sample[col + '_year'] = df_sample[col].dt.year
    df_sample[col + '_month'] = df_sample[col].dt.month
    df_sample[col + '_day'] = df_sample[col].dt.day
df_sample.drop(columns=date_columns, inplace=True)

df_sample = drop_high_missing(df_sample)
df_sample = impute_missing(df_sample)
df_sample = remove_high_corr(df_sample)

num_cols = df_sample.select_dtypes(include=['int64', 'float64']).columns.tolist()
num_cols = [col for col in num_cols if col != 'target']
low_cat, high_cat = filter_categorical_features(df_sample)
low_cat = [col for col in low_cat if col in df_sample.columns]
high_cat = [col for col in high_cat if col in df_sample.columns]

df_sample[low_cat + high_cat] = df_sample[low_cat + high_cat].astype(str)

# Preprocessing Pipeline
num_pipeline = Pipeline([('imputer', SimpleImputer(strategy='median')), ('scaler', StandardScaler())])
low_cat_pipeline = Pipeline([('imputer', SimpleImputer(strategy='constant', fill_value="Missing")), ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))])
high_cat_pipeline = Pipeline([('imputer', SimpleImputer(strategy='constant', fill_value="Missing")), ('freq_enc', FrequencyEncoder())])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_pipeline, num_cols),
        ('low_cat', low_cat_pipeline, low_cat),
        ('high_cat', high_cat_pipeline, high_cat)
    ],
    remainder='drop'
)

# Fit pipelines explicitly
print("Fitting preprocessing pipeline...")
for name, pipeline, columns in preprocessor.transformers:  # Corrected to 'transformers'
    if name != 'remainder':
        pipeline.fit(df_sample[columns])  # Fit each pipeline individually
preprocessor.fit(df_sample)  # Fit the ColumnTransformer
X_sample = preprocessor.transform(df_sample)
y_sample = df_sample['target'].values
print("Sample transformed. Shape:", X_sample.shape)

# Feature Selection with Random Forest
print("Starting feature selection with Random Forest...")
rf = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
rf.fit(X_sample, y_sample)
importances = rf.feature_importances_
feature_names = preprocessor.get_feature_names_out()

# Select top 15 features
n_features_to_select = 15
indices = np.argsort(importances)[::-1][:n_features_to_select]
mask = np.zeros_like(importances, dtype=bool)
mask[indices] = True
selected_feature_names = feature_names[mask]
print("Selected features:", selected_feature_names)
input_dim = len(selected_feature_names)
print(f"Number of features selected: {input_dim}")

# Data Generator with Debugging
def data_generator(file_path, preprocessor, mask, batch_size, is_train=True):
    iterator = pd.read_csv(file_path, chunksize=batch_size, low_memory=False)
    batch_count = 0
    for chunk in iterator:
        print(f"Processing batch {batch_count + 1} of size {len(chunk)}", flush=True)
        chunk['target'] = pd.Categorical(chunk['CASE_STATUS'], categories=categories).codes
        chunk = chunk.drop(columns=['CASE_STATUS', 'CASE_NUMBER'])
        for col in date_columns:
            chunk[col] = pd.to_datetime(chunk[col], errors='coerce')
            chunk[col + '_year'] = chunk[col].dt.year
            chunk[col + '_month'] = chunk[col].dt.month
            chunk[col + '_day'] = chunk[col].dt.day
        chunk.drop(columns=date_columns, inplace=True)
        chunk = impute_missing(chunk)
        for col in low_cat + high_cat:
            if col in chunk.columns:
                chunk[col] = chunk[col].astype(str)
        X_chunk = preprocessor.transform(chunk)
        X_chunk = X_chunk[:, mask]
        y_chunk = chunk['target'].values
        # Pad or truncate to batch_size
        if len(X_chunk) < batch_size:
            pad_size = batch_size - len(X_chunk)
            X_chunk = np.pad(X_chunk, ((0, pad_size), (0, 0)), mode='constant')
            y_chunk = np.pad(y_chunk, (0, pad_size), mode='constant', constant_values=-1)
        elif len(X_chunk) > batch_size:
            X_chunk = X_chunk[:batch_size]
            y_chunk = y_chunk[:batch_size]
        batch_count += 1
        yield X_chunk.astype(np.float32), y_chunk.astype(np.int32)

# Dataset Creation
train_dataset = tf.data.Dataset.from_generator(
    lambda: data_generator(DATA_PATH, preprocessor, mask, BATCH_SIZE, is_train=True),
    output_signature=(
        tf.TensorSpec(shape=(BATCH_SIZE, input_dim), dtype=tf.float32),
        tf.TensorSpec(shape=(BATCH_SIZE,), dtype=tf.int32)
    )
).shuffle(buffer_size=1000).prefetch(tf.data.AUTOTUNE)

val_dataset = tf.data.Dataset.from_generator(
    lambda: data_generator(DATA_PATH, preprocessor, mask, BATCH_SIZE, is_train=False),
    output_signature=(
        tf.TensorSpec(shape=(BATCH_SIZE, input_dim), dtype=tf.float32),
        tf.TensorSpec(shape=(BATCH_SIZE,), dtype=tf.int32)
    )
).prefetch(tf.data.AUTOTUNE)

# ANN Model
def build_ann_model(input_dim, num_classes=4):
    model = keras.Sequential([
        layers.Input(shape=(input_dim,)),
        layers.Dense(512, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        layers.Dense(256, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
        layers.BatchNormalization(),
        layers.Dropout(0.4),
        layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
        layers.BatchNormalization(),
        layers.Dropout(0.3),
        layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
        layers.BatchNormalization(),
        layers.Dropout(0.2),
        layers.Dense(num_classes, activation='softmax')
    ])
    return model

with strategy.scope():
    model = build_ann_model(input_dim)
    model.compile(optimizer=optimizers.Adam(learning_rate=0.0003), loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Training with Debug Steps
checkpoint = keras.callbacks.ModelCheckpoint("model_checkpoint.h5", save_best_only=True)
early_stopping = keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
reduce_lr = keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6)

print("Starting training...")
history = model.fit(
    train_dataset,
    steps_per_epoch=DEBUG_STEPS,  # Use smaller steps for debugging
    epochs=20,
    validation_data=val_dataset,
    validation_steps=VAL_STEPS,
    callbacks=[checkpoint, early_stopping, reduce_lr],
    verbose=1
)

print("Evaluating model...")
loss, accuracy = model.evaluate(val_dataset, steps=VAL_STEPS)
print("Test accuracy:", accuracy)
```
